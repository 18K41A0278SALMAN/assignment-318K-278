3.compute outer product of 2 vectors

```
import numpy as np
r=[[1,0],[0,1]]
s=[[1,2],[3,4]]
print('given arryas are')
print(r)
print(s)
print('outer product is given by=')
p=np.outer(r,s)
print(p)
```

4.write numpy program to compute the condition number of a given matrix .original matrix

```
import numpy as np
from numpy import linalg as la
a=np.array([[1,2],[3,4]])
print(la.cond(a))
```

5. write a NumPy program to compute the sum of the diagonal element of a given array(matrix).

```
x=np.arange(9).reshape(3,3)
print(x)
s=np.trace(x)
print(s)
```
